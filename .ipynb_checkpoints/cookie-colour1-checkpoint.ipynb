```python
import PIL
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
img = PIL.Image.open('cookie0.jpg') 
img1 = cv.imread('cookie0.jpg', 1)
```

```python
img
```

```python
#scale the image
scaled_cookie = cv.resize(img1, (255,255))
```

```python
plt.imshow(scaled_cookie)
```

```python
scaled_cookie.shape
```

```python
#grabcut for extracting the cookie foreground
mask = np.zeros(scaled_cookie.shape[:2], np.uint8)

bgmodel = np.zeros((1,65), np.float64)
fgmodel = np.zeros((1,65), np.float64)

#only the edges(10 px) will def be the background 
rectangle = (10, 10, 244, 244) 

cv.grabCut(scaled_cookie, mask, rectangle, bgmodel, fgmodel, 3, cv.GC_INIT_WITH_RECT)

mask2 = np.where((mask == 2)|(mask == 0), 0, 1).astype('uint8') 

image_fg = scaled_cookie * mask2[:, :, np.newaxis] 
```

```python
plt.imshow(image_fg)
```

```python
image_fg.shape
```

```python
rgb_list = []
for i in range(image_fg.shape[0]):
    for j in range(image_fg.shape[1]):
        rgb_list.append(image_fg[i][j]) #storing rbg values for all the pixels
        
def rgb_to_hex(rgb):
    rgb = (rgb[0], rgb[1], rgb[2])
    return '%02x%02x%02x' % rgb

hex_list = []
for pixel in rgb_list: #convert each rbg value to hex
    hex_list.append(rgb_to_hex(pixel))
    
import pandas as pd 
hex_counts = pd.Series(hex_list).value_counts() # gives counts for each hex value

def hex_to_rgb(value):
    value = value.lstrip('#')
    lv = len(value)
    return tuple(int(value[i:i+lv//3], 16) for i in range(0, lv, lv//3))


print("most occuring hex values")
print(str(hex_counts.index[0]))
print(str(hex_counts.index[1]))
print(str(hex_counts.index[2]))
print(str(hex_counts.index[3]))
```

```python
pix = np.array([hex_to_rgb(hex_counts.index[1]), hex_to_rgb(hex_counts.index[2]), hex_to_rgb(hex_counts.index[3])]) 
```

```python
plt.imshow(pix, cmap = 'gray')
```

```python
cv.imwrite("./pix.jpg", pix)
```

```python
#colors of only the cookie
only_cookie = []

for row in range(10, image_fg.shape[0]-10):
    for col in range(10, image_fg.shape[1]-10):
        
        if(sum(image_fg[row,col]) != 0):
            #print(image_fg[row,col])
            only_cookie.append(list(image_fg[row,col]))
```

```python
len(only_cookie)
```

```python
#averaging the reds, blues and greens to get the overall average colour
b_val = 0
g_val = 0
r_val = 0

for r,g,b in only_cookie:
    b_val += b
    g_val += g
    r_val += r

avg_color = [b_val, g_val, r_val]
```

```python
avg_color = [math.floor(val/len(only_cookie)) for val in avg_color]
```

```python
cookie_color_samp = np.array([[avg_color] * 100]*100)
```

```python
plt.imshow(cookie_color_samp)
```
